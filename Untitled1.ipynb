# METAL STORM


Cilj tega projekta je predstaviti podatke, vzete iz spletne strani http://www.metalstorm.net. Dobljeni podatki obsegajo najboljših 100 (ali manj) albumov, ki so na strani ocenjeni za najboljše albume posameznih podrzvrsti. Vsak album vsebuje sledeče podatke: 
* avtor
* leto izdaje
* podzvrst, ki ji pripada
* ocena 
* število glasov
* mesto na lestvici posamezne podrzvrsti

Že na sami strani so bili albumi do neke mere filtrirani, saj je pogoj za uvrstitev na lestvico vsaj 20 glasov. Lestvica prav tako ne vsebuje EP-jev, Live albumov in kompilacij, torej zgolj LP izdaje.

Najprej si za predstavo oglejmo celotno tabelo z vsemi zbranimi podatki iz strani in importajmo matplotlib, ki ga bomo potrebovali kasneje:

```python
import pandas as pd
%matplotlib
podatki = pd.read_csv('podatki.csv')
podatki
```

Občasno bom izkoristil še funkcijo, ki napiše celoten seznam, saj nas ta občasno zanima bolj kot le skrajšana verzija

```python
def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')
```

Ena izmed reči, ki jih prav tako želim izvedeti je: v katerih letih je bilo znotraj določene podzvrsti izdanih največ albumov iz teh seznamov, oziroma je ta podzvrst doživela najvelčji porast. Oglejmo si graf:

7 najbolje ocenjenih albumov vseh časov:

```python
podatki[podatki.ocena > 9.10].sort_values(by = ['ocena'], ascending = False)
```

Leta 2007 je bilo torej od vseh let največ albumov, ki so prišli na to lestvico.

```python
podatki.groupby(['leto','podzvrst']).size().unstack()
```

```python
podatki.groupby(['leto','podzvrst']).size().unstack()\
.apply(lambda x: x / x.sum(), axis=1)\
.plot(kind='area', stacked=False, table=True)
```

```python
podatki.groupby(['leto','podzvrst']).size().unstack()\
.apply(lambda x: x / x.sum(), axis=1)\
.plot(kind='area', table=True)
```
